# Нормальное/гауссово распределение

__Стандартное/нормальное распределение__ $\mathcal N(0,1)$ имеет следующие функции плотности и распределения

$$
\begin{aligned}
	\phi(x)&=\frac{1}{\sqrt{2\pi}}\exp\left(-\frac{x^2}{2}\right) &
	\Phi(x)&=\int_{-\infty}^x \phi(t)dt & x&\in\mathbb{R}
\end{aligned}
$$

__Общее стандартное/нормальное распределение__ $\mathcal N(\mu,\sigma^2)$ с параметрами 

|Параметр|назначение|
|-|-|
$\mu\in\mathbb{R}$ | параметр сдвига|
|$\sigma>0$| параметр масштабаЁ

имеет следующие функции плотности и распределения

$$
\begin{aligned}
	f(x)&=\frac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)=\frac{1}{\sigma}\phi\left(\frac{x-\mu}{\sigma}\right) &
	F(x)&=\int_{-\infty}^x f(t)dt=\Phi\left(\frac{x-\mu}{\sigma}\right) & x&\in\mathbb{R}
\end{aligned}
$$


Пусть $X\sim \mathcal N(\mu,\sigma^2)$. Тогда

$$
\begin{aligned}
	E(X)&=\mu & Var(X)&=\sigma^2 & \sigma(X)&=\sigma
\end{aligned}
$$

```python
import numpy as np
from scipy.stats import norm

import matplotlib.pyplot as plt
```

## Плотность и функция распределения

```python
mu1, mu2, mu3 = 0, 0, 0
sd1, sd2, sd3 = 1, 0.5, 2
x_min, x_max = -5, 5

rv1, rv2, rv3 = norm(mu1, sd1), norm(mu2, sd2), norm(mu3, sd3)

x = np.linspace(start=x_min, stop=x_max, num=100)

fig, axs =  plt.subplots(nrows=1, ncols=2, figsize=(12, 8))

axs[0].plot(x, rv1.pdf(x), label='('+str(mu1)+','+ str(sd1)+'^2)')
axs[0].plot(x, rv2.pdf(x), label='('+str(mu2)+','+ str(sd2)+'^2)')
axs[0].plot(x, rv3.pdf(x), label='('+str(mu3)+','+ str(sd3)+'^2)')

axs[1].plot(x, rv1.cdf(x), label='('+str(mu1)+','+ str(sd1)+'^2)')
axs[1].plot(x, rv2.cdf(x), label='('+str(mu2)+','+ str(sd2)+'^2)')
axs[1].plot(x, rv3.cdf(x), label='('+str(mu3)+','+ str(sd3)+'^2)')

axs[0].set_title('Плотность')
axs[1].set_title('Функция распределения')

axs[0].legend()
axs[1].legend()

axs[0].hlines(0, x_min, x_max, linestyles='dotted')
axs[1].hlines(0, x_min, x_max, linestyles='dotted')
axs[1].hlines(1, x_min, x_max, linestyles='dotted')
axs[1].vlines(0, 0, 1, linestyles='dotted')

plt.show()
```

## Вычисление вероятностей

Пусть $X\sim \mathcal N(1, 2^2)$. Вычислим 

$$
\begin{aligned}
	&P(X\leq 2) & &P(X>1.5) & &P(0<X\leq 2)
\end{aligned}
$$

```python
norm(1, 2).cdf(2).round(3), norm(1, 2).sf(1.5).round(3), (norm(1, 2).cdf(2)-norm(1, 2).cdf(0)).round(3)
```
